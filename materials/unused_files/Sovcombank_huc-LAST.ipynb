```python
# usd
url = "https://api.apilayer.com/exchangerates_data/timeseries?start_date=2022-10-20&end_date=2022-11-20&symbols=RUB&base=USD" # &symbols=RUB,USD

payload = {}
headers= {
  "apikey": "JJNdsqUntjo5Q5yjQNAOR353F7D6rlxJ"
}

response = requests.request("GET", url, headers=headers, data = payload)

status_code = response.status_code
result = response.text
usd = response.json()

usd = pd.DataFrame(usd).reset_index()
usd = pd.json_normalize(usd["rates"]).unstack().apply(pd.Series).droplevel(level=0).join(usd["index"]).rename(columns={0: 'curs_usd', 'index': 'date'})
usd.plot(
    x='date', y='curs_usd', kind='line', style='-', linewidth=3.5, grid=True, figsize=(15, 7), 
    title='USD', ylabel='Курс доллара', xlabel='Дата', color='purple')
plt.show()
```
